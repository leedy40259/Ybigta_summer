```python
def strength(alcohol):
  if alcohol > 10:
    return 'strong'
  else: 
    return 'weak'

data['strength'] = data['alcohol'].apply(strength)
data.head(30)
```

```python
len(df_y)
```

#### df_x와 df_y를 inner join시켜라 (quality를 기준으로)

```python
pd.merge(df_x, df_y, how = 'inner', on = 'quality')
```

#### df_x에서 df_y를 left join시켜라 (quality를 기준으로)

```python
pd.merge(df_x, df_y, how = 'left', on = 'quality')
```

```python
# '날짜' 컬럼을 보고, '월' 컬럼과 '계절' 컬럼을 새로 추가하라. 
# (계절 컬럼 : 3-5월 : 봄, 6-8월 : 여름, 9-11월: 가을, 12-2월 : 겨울)

# '날짜' 컬럼의 dtype 확인
deliver[['날짜']].info()
```

```python
# 날짜에서 월 추출하기
import datetime

# 월 함수
def month(date):
  datem = datetime.datetime.strptime(date, "%Y-%m-%d")
  return datem.month

deliver['월'] = deliver['날짜'].apply(month)
deliver.head()
```

```python
# 계절 함수
def season(month):
  if month >= 3 and month <=5:
    return '봄'
  elif month >= 6 and month <= 8:
    return '여름'
  elif month >= 9 and month <= 11:
    return '가을'
  else:  
    return '겨울'

deliver['계절'] = deliver['월'].apply(season)
deliver.head()
```

```python
# '미세먼지(㎍/㎥)' 컬럼에서 null값의 개수를 확인하고, null값을 전체 평균으로 대체하라.
# (전체 평균은 null값이 아닌 값들의 평균으로 계산된다.)

# null값 개수 확인
deliver['미세먼지(㎍/㎥)'].isnull().sum()
```

```python
#null값을 전체 평균으로 대체
sample = deliver['미세먼지(㎍/㎥)'].dropna()
deliver['미세먼지(㎍/㎥)'] = deliver['미세먼지(㎍/㎥)'].fillna(sample.mean())
# deliver['미세먼지(㎍/㎥)'].fillna(deliver['미세먼지(㎍/㎥)'].mean()) 과 결과는 동일함
deliver.head()
```
